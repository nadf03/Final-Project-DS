```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
data = {
    'TransactionAmount': [
        100, 150, 50, 100, 130, 120, 100, 85, 70, 150, 
        150, 120, 50, 100, 100, 140, 90, 150, 50, 90, 
        120, 100, 110, 75, 65
    ]
}

df = pd.DataFrame(data)
```

```python
# Mengubah data menjadi array
data_array = np.array(df['TransactionAmount'])
```

```python
#Central Tendency
import statistics as statss
mean = statss.mean(data_array)
median = statss.median(data_array)
# Menampilkan hasil
print(f"Mean: {mean}")
print(f"Median: {median}")
```

```python
# Visualisasi Mean dan Median dengan Histogram
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=10, kde=True, color='skyblue')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.legend()
plt.title('Histogram of Transactions with Mean and Median')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

Kesimpulan : 
Hasil analisis menunjukkan bahwa ada variansi yang signifikan dalam transkasi. Walaupun nilai mean berada di angka $100, nilai varians yang tinggi dan jangkauan (range) yang luas menunjukkan tidak semua transaksi berkinerja sama. berdasarkan sebaran ini, perusahaan disarankan untuk mengidentifikasi lebih lanjut faktor-faktor yang menyebabkan transaksi renda, serta melihat segmen pelanggan yang melakukan transaksi tinggi untuk memperbaiki dan memahami pola pembelian pelanggan

```python
#Define alpha = 0.05
alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test
t_statistic, p_value = stats.ttest_1samp(a=data_array, popmean=100)

# Menampilkan hasil
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print("Reject H0:", p_value < alpha)
```

#Conclude with alpha and p-value
Nilai p-value lebih besar dari alpha = 0.05, artinya gagal untuk tolak H0 atau terima H0

Conclusion: Hasil ini menunjukkan bahwa pelatihan yang dilakukan tidak memberikan efek signifikan pada peningkatan rata-rata penjualan. Dengan kata lain, rata-rata penjualan setelah pelatihan tidak berbeda secara signifikan dari rata-rata penjualan sebelumnya, yaitu $100.

```python
#Conclude with t statistics and critical region
#Find the critical region
```

```python
#H1
n = len(data_array)  # Sample size
df = n - 1
#Find critical t-value for one-tailed test
critical_t = stats.t.ppf(1 - alpha, df)
# Define critical region
critical_region = (critical_t, np.inf)  # Since it's a one-tailed test
print(f"Critical T-Value: {critical_t}")
print(f"Critical Region: {critical_region}")
```
